```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
model_name = "llama3"
baseline_df = pd.read_csv(f"/home/max/Studium/Leipzig/Semester6/Math_and_ML/math-ml/results/accuracy_exp/all_baseline_{model_name}/evaluation_results.csv")
cot_df = pd.read_csv(f"/home/max/Studium/Leipzig/Semester6/Math_and_ML/math-ml/results/accuracy_exp/all_cot_{model_name}/evaluation_results.csv")
```

```python
#remove buggy samples
baseline_data = baseline_df[["entropy", "cosine", "correct"]]
baseline_data = baseline_data[baseline_data['correct'] != 'buggy']

cot_data = cot_df[["entropy", "cosine", "correct"]]
cot_data = cot_data[cot_data['correct'] != 'buggy']
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(baseline_data["entropy"], baseline_data["cosine"], alpha=0.6, label=f"{model_name} Baseline", marker='o')
plt.scatter(cot_data["entropy"], cot_data["cosine"], alpha=0.6, label=f"{model_name} CoT", marker='o')
plt.xlabel("Entropy")
plt.ylabel("Cosine Similarity")
plt.title("Entropy vs Cosine Similarity: Baseline vs CoT")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{model_name}.png")
```

```python
import matplotlib.pyplot as plt

# Data
tokens = ["7", "624", "six", "6240", "5"]
probs = [0.02, 0.8, 0.03, 0.14, 0.01]

# Plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(tokens, probs)

# Labels
ax.set_ylabel("token probabilty")

# Clean look
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
fig.tight_layout()

# Save figure for download
file_path = "show_high_entropy.png"
fig.savefig(file_path, dpi=300)
```

```python
import matplotlib.pyplot as plt

# 2-D toy coordinates chosen to be very close (high cosine similarity)
coords = {
    "78.5": (0.8, 0.8),
    "seventy-eight": (0.8, 0.7)
}

fig, ax = plt.subplots(figsize=(4, 4))

# Draw vectors (lines from origin) and scatter points
for token, (x, y) in coords.items():
    ax.plot([0, x], [0, y])  # vector line
    ax.scatter(x, y)          # point
    ax.text(x + 0.02, y, token)  # label slightly offset

# Aesthetic tweaks: keep it simple and clean
ax.set_xlabel("dimension 1")
ax.set_ylabel("dimension 2")
ax.set_aspect("equal", adjustable="box")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
fig.tight_layout()

# Save figure
file_path = "cosine_sim.png"
fig.savefig(file_path, dpi=300)
```
