```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
model_name = "mistral"
prompting_technique = "cot"
df = pd.read_csv(f"/home/julian/dev/math-ml/results/accuracy_exp/all_{prompting_technique}_{model_name}/evaluation_results.csv")
```

```python
#remove buggy samples
data = df[["entropy", "cosine", "correct"]]
correct_data = data[data['correct'] == 'yes']
incorrect_data = data[data['correct'] == 'no']
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(correct_data["entropy"], correct_data["cosine"], alpha=0.6, label=f"{model_name}  Correct", marker='o')
plt.scatter(incorrect_data["entropy"], incorrect_data["cosine"], alpha=0.6, label=f"{model_name} Incorrect", marker='o')
plt.xlabel("Entropy")
plt.ylabel("Cosine Similarity")
plt.title("Entropy vs Cosine Similarity: Correct vs Incorrect")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{model_name}_{prompting_technique}_accuracy.png")
```
