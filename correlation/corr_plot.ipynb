```python
import matplotlib.pyplot as plt
from collections import defaultdict
# Data: (Model, Pearson r, Spearman r)

data = [
    ("Qwen3-8B", "gsm8k", -0.37, -0.39),
    ("Qwen3-8B", "writingprompts", -0.38, -0.41),
    ("Qwen3-8B", "xsum", -0.37, -0.43),

    ("Mistral-7B-v0.1", "gsm8k", -0.50, -0.67),
    ("Mistral-7B-v0.1", "writingprompts", -0.49, -0.66),
    ("Mistral-7B-v0.1", "xsum", -0.51, -0.67),

    ("Llama-3.1-8B-Instruct", "gsm8k", -0.57, -0.70),
    ("Llama-3.1-8B-Instruct", "writingprompts", -0.58, -0.70),
    ("Llama-3.1-8B-Instruct", "xsum", -0.57, -0.69),

    ("Llama-3.1-8B", "gsm8k", -0.62, -0.76),
    ("Llama-3.1-8B", "writingprompts", -0.61, -0.75),
    ("Llama-3.1-8B", "xsum", -0.63, -0.76),

    ("deepseek-llm-7b-base", "gsm8k", -0.42, -0.65),
    ("deepseek-llm-7b-base", "writingprompts", -0.43, -0.67),
    ("deepseek-llm-7b-base", "xsum", -0.45, -0.67),
]


grouped = defaultdict(lambda: {"x": [], "y": []})
for model, dataset, p, s in data:
    grouped[model]["x"].append(p)
    grouped[model]["y"].append(s)

# Plot
plt.figure(figsize=(7, 7))
for model, values in grouped.items():
    plt.scatter(values["x"], values["y"], label=model)

plt.xlabel("Pearson r")
plt.ylabel("Spearman r")
plt.title("Cosine Correlation Scores by Model and Dataset")
plt.xlim(0, -1)
plt.ylim(0, -1)
plt.grid(True)
plt.legend(title="Model")
plt.tight_layout()
plt.savefig("../results/correlation_eval/corr_scores.png")
plt.show()
```
