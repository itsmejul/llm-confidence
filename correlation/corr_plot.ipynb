```python
import matplotlib.pyplot as plt

# Data: (Model, Pearson r, Spearman r)

data = [
    ("Qwen3", "gsm8k", -0.37, -0.39),
    ("Qwen3-8B", "writingprompts", -0.38, -0.41),
    ("Qwen3-8B", "xsum", -0.37, -0.43),

    ("Mistral", "gsm8k", -0.50, -0.67),
    ("Mistral-7B-v0.1", "writingprompts", -0.49, -0.66),
    ("Mistral-7B-v0.1", "xsum", -0.51, -0.67),

    ("Llama-Instruct", "gsm8k", -0.57, -0.70),
    ("Llama-3.1-8B-Instruct", "writingprompts", -0.58, -0.70),
    ("Llama-3.1-8B-Instruct", "xsum", -0.57, -0.69),

    ("Llama", "gsm8k", -0.62, -0.76),
    ("Llama-3.1-8B", "writingprompts", -0.61, -0.75),
    ("Llama-3.1-8B", "xsum", -0.63, -0.76),

    ("Deepseek", "gsm8k", -0.42, -0.65),
    ("deepseek-llm-7b-base", "writingprompts", -0.43, -0.67),
    ("deepseek-llm-7b-base", "xsum", -0.45, -0.67),
]
# Unpack data
models, datasets, pearson_r, spearman_r = zip(*data)

# Create scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(pearson_r, spearman_r)

# Add labels to each point
for i, model in enumerate(models):
    if i% 3 == 0:
        plt.text(pearson_r[i], spearman_r[i], model, fontsize=9, ha='right')
        plt.text(pearson_r[i], spearman_r[i], "", fontsize=9, ha='right')

# Set axis labels
plt.xlabel("Pearson r")
plt.ylabel("Spearman r")
plt.title("Correlation Scores on GSM8K")

# Set axis limits and invert so that 0 is at the origin
plt.xlim(0, -1)
plt.ylim(0, -1)
plt.grid(True)
plt.tight_layout()
plt.show()
```
