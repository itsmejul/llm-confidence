```python
import matplotlib.pyplot as plt

#What is 7 times 8?

# Define token labels and probabilities for high and low entropy distributions
tokens1 = ['56', '5', 'five', 'fifty-', 'fifteen']
tokens2 = ['56', '5', 'five', 'fifty-', 'fifteen']

# Low entropy: one dominant token
low_entropy_probs = [0.7, 0.15, 0.09, 0.03, 0.01]

# High entropy: more uniform distribution
high_entropy_probs = [0.23, 0.15, 0.14, 0.21, 0.27]

# Plot low entropy distribution
plt.figure(figsize=(6, 4))
plt.bar(tokens1, low_entropy_probs)
plt.title('Low Entropy Token Distribution')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()

# Plot high entropy distribution
plt.figure(figsize=(6, 4))
plt.bar(tokens2, high_entropy_probs)
plt.title('High Entropy Token Distribution')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```
